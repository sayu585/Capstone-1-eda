### Variables Description

---
**hotel**: The type of hotel, indicating whether it is a city hotel or a resort hotel.

**is_canceled**: A binary variable, where 1 indicates that the booking was canceled and 0 indicates that the booking was not canceled.

**lead_time**: The number of days between the booking date and the arrival date.

**arrival_date_year**: The year of the arrival date.

**arrival_date_month**: The month of the arrival date.

**arrival_date_week_number**: The week number of the year for the arrival date.

**arrival_date_day_of_month**: The day of the month for the arrival date.

**stays_in_weekend_nights**: The number of weekend nights (Saturday or Sunday) stayed.

**stays_in_week_nights**: The number of week nights (Monday to Friday) stayed.

**adults**: The number of adults included in the booking.

**children**: The number of children included in the booking.

**babies**: The number of babies included in the booking.

**meal**: The type of meal booked, such as Bed & Breakfast, Half Board, Full Board, etc.

**country**: The country of origin for the guests.

**market_segment**: The market segment designation, representing the customer segment (e.g., Online TA, Offline TA/TO, Direct, etc.).

**distribution_channel**: The booking distribution channel, indicating how the booking was made (e.g., Direct, Corporate, Travel Agents, etc.).

**is_repeated_guest**: A binary variable indicating whether the guest is a repeated guest (1) or not (0).

**previous_cancellations**: The number of previous booking cancellations by the guest.

**previous_bookings_not_canceled**: The number of previous bookings that were not canceled by the guest.

**reserved_room_type**: The type of room that was reserved.

**assigned_room_type**: The type of room that was assigned at check-in.

**booking_changes**: The number of changes made to the booking.

**deposit_type**: The type of deposit made for the booking (e.g., No Deposit, Non-Refundable, Refundable).

**agent**: The ID of the travel agency that made the booking (if applicable).

**company**: The ID of the company/entity that made the booking (if applicable).

**days_in_waiting_list**: The number of days the booking was on the waiting list before it was confirmed.

**customer_type**: The type of customer (e.g., Transient, Contract, Group, etc.).

**adr**: The Average Daily Rate, calculated as the sum of all room nights divided by the number of nights stayed.

**required_car_parking_spaces**: The number of car parking spaces required by the guest.

**total_of_special_requests**: The total number of special requests made by the guest.

**reservation_status**: The current status of the reservation (e.g., Check-Out, Canceled, No-Show).

**reservation_status_date**: The date when the reservation status was last updated.

### **Eliminate duplicate row**

````python
# Dropping duplicate values
htlcpy.drop_duplicates(inplace = True)
````

**Handling missing values**

````python
# Columns having missing values
htlcpy.isnull().sum().sort_values(ascending = False)[:10]
````

Since the data in the company and agent columns includes the company number and agency number. There may be instances where a customer does not book a hotel through an agent or a company. In that situation, values in these columns can be null. In these columns, null values will be changed to zero.

````python
htlcpy[['company','agent']] = htlcpy[['company','agent']].fillna(0)
````

````python
htlcpy['children'].unique()
````

This column's value of 0 indicates that there were no children among the clients who completed the transaction. Therefore, "nan" values are missing values as a result of data recording errors. The children's mean value will be used to replace the null values under this column.

````python
htlcpy['children'].fillna(htlcpy['children'].mean(), inplace = True)
````

The next column, labelled "country," is empty. The customer's country of origin is represented in this column. Considering that this column's datatype is a string. We will replace the missing value with the mode of 'country' column.

````python
htlcpy['country'].fillna('others', inplace = True)
````

````python
# Checking if all null values are removed
htlcpy.isnull().sum().sort_values(ascending = False)[:10]
````

There are some rows with a total adult, kid, or baby count of 0. So, we'll eliminate of those rows.

````python
htlcpy[htlcpy['adults']+htlcpy['babies']+htlcpy['children'] == 0].shape
````

````python
htlcpy.drop(htlcpy[htlcpy['adults']+htlcpy['babies']+htlcpy['children'] == 0].index,inplace = True)
````

````python
htlcpy.shape
````

**Converting columns to appropriate datatypes**

````python
# Converting datatype of columns 'children', 'company' and 'agent' from float to int.
htlcpy[['children', 'company', 'agent']] = htlcpy[['children', 'company', 'agent']].astype('int64')
````

````python
# changing datatype of column 'reservation_status_date' to data_type.
htlcpy['reservation_status_date'] = pd.to_datetime(htlcpy['reservation_status_date'], format = '%Y-%m-%d')
````

**Adding important columns**

````python
# Adding total staying days in hotels
htlcpy['total_stay'] = htlcpy['stays_in_weekend_nights']+htlcpy['stays_in_week_nights']
````

````python
# Adding total people num as column, i.e. total people num = num of adults + children + babies
htlcpy['total_people'] = htlcpy['adults']+htlcpy['children']+htlcpy['babies']
````

This column is being added so that we can analyse how long people stay in hotels.
